```python
import numpy as np
import pandas as pd
```

### Experiment 1 Fitbit charge

Read the CSV 

```python
hr = pd.read_csv("mmc2.csv")
tweet = pd.read_csv("mmc3.csv")
```

```python
hr = hr.dropna()
```

```python
hr.info()
```

```python
tweet.info()
```

Change data format

```python
hr["date_time"] = pd.to_datetime(hr["date_time"], format="%Y-%m-%d %H:%M:%S")
tweet["date_time"] = pd.to_datetime(tweet["date_time"], format="%Y-%m-%d %H:%M:%S")
```

Calculate minute difference from start of experiment to corelate tweet to HR

```python
hr["hrdiff"] = (hr["date_time"] - pd.Timestamp('2016-05-10 00:00:00')).astype('timedelta64[m]')
tweet["twdiff"] = (tweet["date_time"] - pd.Timestamp('2016-05-10 00:00:00')).astype('timedelta64[m]')
```

```python
print(hr)
```

```python
print(tweet)
```

```python
df2 = pd.merge(hr, tweet, how='inner', left_on='hrdiff', right_on='twdiff')
```

```python
df2.info()
```

```python
print(df2)
```

```python
result = df2[['hr','sent_num']].copy()
```

```python
result
```

```python
result.to_csv("result_ofitbit.csv", index= False)
```
